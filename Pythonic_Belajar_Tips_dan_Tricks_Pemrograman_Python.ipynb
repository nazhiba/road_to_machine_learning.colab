### Code

```python
x = 10
y = 20
print("Sebelum pertukaran")
print(f"x = {x} y = {y}")
z = x # Variabel baru bernilai X, Z = 10
x = y # X = 20
y = z # Y = 10
print("Setelah pertukaran")
print(f"x = {x} y = {y}")


# CARA YANG LEBIH PYTHONIC
print("\nCara Lebih Pythonic")
x,y = 10,20 # Ora meng 2 Vairabel. iso luweh x,y,z,... (dicobo deweðŸ’€)
print("Sebelum pertukaran")
print(f"x = {x} y = {y}")
x,y = y,x
print("Setelah pertukaran")
print(f"x = {x} y = {y}")
```

### Code

```python
# Cara Tidak Pythonic
print("Cara Tidak Pythonic")
x = 10
y = 20
z = 30
t = [None]*3
t[0] = x # Array di Python dimulai dari 0 bukan 1...dst
t[1] = y
t[2] = z
print(f"x = {x} y = {y} z = {z}")
print(f"t = {t}")

# Cara Lebih Pythonic
print("\nCara Lebih Pythonic")
x,y,z = 10,20,30
t = [x,y,z]
print(f"x = {x} y = {y} z = {z}")
print(f"t = {t}")
```

```python
# Cara Tidak Pythonic
print("Cara Tidak Pythonic")
listnaz = [10,20,30]
x = listnaz[0]
y = listnaz[1]
z = listnaz[2]
print(f"listnaz = {listnaz}")
print(f"x = {x} y = {y} z = {z}")
# Cara Lebih Pythonic
print("\nCara Lebih Pythonic")
listnaz = [10,20,30]
x,y,z = listnaz
print(f"listnaz = {listnaz}")
print(f"x = {x} y = {y} z = {z}")
```

```python
# Cara Tidak Pythonic
print("Cara Tidak Pythonic")
listnaz = [10,20,30]
for i in range(len(listnaz)):
  print(f"{listnaz[i]}")

# Cara Lebih Pythonic
print("\nCara Lebih Pythonic")
for i in listnaz:
  print(f"{i}")
```

```python
# Cara Tidak Pythonic
print("Cara Tidak Pythonic")
nik = ["0882","0853","0891"]
jeneng = ["Anes","Prbawo","Gahar"]
for i in range(len(nik)):
  print(f"{nik[i]} --- {jeneng[i]}")

# Cara Lebih Pythonic
print("\nCara Lebih Pythonic")
for i,data_nik in enumerate(nik):
  print(f"{data_nik} ... {jeneng[i]}")

```
